```
import numpy as np
import matplotlib.pyplot as plt  
from sklearn.decomposition import PCA


# total number of classes
total_classes = 345
# at every subplot we cpmpare 2 classes
num_classes_for_pca = 2
# we draw principal components for 500 samples from every class
num_samples = 500
# we draw 9 subplots, 3x3
height_subplots = 3
width_subplots = 3

# We want the 2 principal components for the visualization
pca = PCA(n_components=2)  

# subplot initialization
fig, ax = plt.subplots(height_subplots, width_subplots, sharex='col', sharey='row',figsize=(15,15))

# drawing each subplot
for k in xrange(height_subplots):
  for j in xrange(width_subplots):
    pca_class_ids = np.random.randint(0,total_classes,num_classes_for_pca)
    full_data = []
    num_examples = []
    for i in pca_class_ids:
      reader = '/Users/shiva/Downloads/'+ list_names[i]+'.npy'
      this_class_data = np.load(reader)
      full_data.append(this_class_data[0:num_samples,:])
      num_examples.append(this_class_data.shape[0])

    full_data = np.vstack(full_data)
    principalComponents = pca.fit_transform(full_data)

    ax[k,j].scatter(principalComponents[0:num_samples,0], principalComponents[0:num_samples,1], s=15, c='b', marker="s", label=list_names[pca_class_ids[0]])
    ax[k,j].scatter(principalComponents[num_samples:,0],principalComponents[num_samples:,1], s=15, c='r', marker="s", label=list_names[pca_class_ids[1]])
    ax[k,j].set_ylabel('PC2')
    ax[k,j].set_xlabel('PC1')
    ax[k,j].set_yticklabels([])
    ax[k,j].set_xticklabels([])
    ax[k,j].legend(loc='upper left')

plt.show()
```
